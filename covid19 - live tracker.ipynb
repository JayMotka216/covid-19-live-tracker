# COVID19 India Live tracker

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read data from goverment site
data = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
data
```

```python
data = data.iloc[:-1,1:]
data.tail()
```

```python
total_state = np.arange(len(data['state_name']))
total_state
```

```python
data['active']
```

```python
max(data['positive'])
```

## Total Positive Case based on states in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 10), dpi=70, facecolor='w', edgecolor='y')
plt.barh(total_state,data['positive'], align='center', alpha=0.5,  
                 color=(0,0.5,1),  
                 edgecolor=(0.5,0.2,1))
    
plt.yticks(total_state, data['state_name'],fontsize=14)  
plt.xlim(1,max(data['positive'])+100000) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

## Active Number of cases based on state

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 10), dpi=70, facecolor='w', edgecolor='y')
plt.barh(total_state,data['active'], align='center', alpha=0.5,  
                 color=(1,0,0.01),  
                 edgecolor=(0.5,0.2,1))
    
plt.yticks(total_state, data['state_name'],fontsize=14)  
plt.xlim(1,max(data['active'])+10000) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

## Total diseased based on state in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(10,10), dpi=80, facecolor='w', edgecolor='y' )
plt.barh(total_state, data['death'], align='center', alpha=0.5, color=(1,0,0), edgecolor=(0.5,0.2,1))

plt.yticks(total_state,data['state_name'],fontsize=12)
plt.xlim(1,max(data['death'])+100)
plt.xlabel('Total deaths')
plt.title('Corona Virus Cases')
plt.show()
```
